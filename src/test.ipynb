```python
import matplotlib.pyplot as plt

# Step 1: Open the file
with open("data_bounce.txt", "r") as file:
    # Step 2: Read the contents line by line
    lines = file.readlines()

# Step 3: Convert each line to a float number
data = [float(line.strip()) for line in lines]

# Step 5: Plot the data
plt.plot(data)
# plt.plot(data)
plt.xlabel("Line Number")
plt.ylabel("Data")
plt.title("Data Plot")
plt.show()
```

```python
import numpy as np

# Step 1: Perform FFT on the data
fft_data = np.fft.fft(data)

# Step 2: Calculate the frequency values
n = len(data)
freq = np.fft.fftfreq(n, d=300e-6)

# Step 3: Plot the FFT results
plt.plot(freq, np.abs(fft_data))
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("FFT Plot")
# plt.xlim(-1, 200)
plt.ylim(0, 100000)
plt.show()
```
